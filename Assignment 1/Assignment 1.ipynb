# Problem 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = plt.imread('washdc512.jpg')
m,n = x.shape
print('imsize(m,n) = ', x.shape)
plt.imshow(x, cmap = 'gray')

mx = x[0,0]
mn = x[0,0]

for i in range(1,m):
    for j in range(1,n):
        if x[i,j] > mx:
            mx = x[i,j]
        if x[i,j] < mn:
            mn = x[i,j]

print('max = ', mx, 'min = ', mn)
```

# Problem 2
### Pixel range is from 0 - 255
### The entirity of dynamic range of brightness values are not being used
### Dynamic range to change from 0 - 95 to 0 - 255
### x[i,j] = (x[i,j] - min) * 255/(mx - mn)

```python
def minmax(x):
    m,n = x.shape
    mx = x[0,0]
    mn = x[0,0]
    for i in range(1,m):
        for j in range(1,n):
            if x[i,j] > mx:
                mx = x[i,j]
            if x[i,j] < mn:
                mn = x[i,j]
    return mx,mn
print('max = ', mx, 'min = ', mn)
```

```python
def drange(x):
    m,n = x.shape
    x1 = np.zeros(shape = (m,n), dtype = 'uint8')
    for i in range(m):
        for j in range(n):
            x1[i,j] = np.uint8((x[i,j] - mn) * 255/(mx - mn))
    return x1
```

```python
def main():
    x = plt.imread('washdc512.jpg')
    print('max, min = ', minmax(x))
    x1 = drange(x)
    print('new image max, min = ', minmax(x1))
    plt.imsave('newwash.jpg', x1, cmap = 'gray')
    plt.imshow(x1, cmap = 'gray')
    
if __name__ == '__main__':
    main()
```

## 2.	The intensity range can be mapped to full range (0-255) by a simple mapping function (H) which maps the input pixel intensity min to 0 and max to 255. Derive the mapping function and implement using python arrays and for loops.


```python
import numpy as np
import matplotlib.pyplot as plt
x = plt.imread('washdc512.jpg')
m,n = x.shape
print('imsize(m,n) = ', x.shape)
plt.imshow(x, cmap = 'gray')

mx = x[0,0]
mn = x[0,0]

def minmax(x):
    m,n = x.shape
    for i in range(1,m):
        for j in range(1,n):
            if x[i,j] > mx:
                mx = x[i,j]
            if x[i,j] < mn:
                mn = x[i,j]
    return mx,mn

print('max = ', mx, 'min = ', mn)
```
